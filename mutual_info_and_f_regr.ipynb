```python
import numpy as np
import pandas as pd 
```

```python
seed=42
np.random.seed(seed)
import os
dir_path='C:\\Data Science\\kaggle regr competion'
csv_file='preprocessedtrain.csv'
dataset=pd.read_csv(os.path.join(dir_path,csv_file))
```

```python
dataset.head(10)
```

```python
X=dataset.loc[:, dataset.columns != 'SalePrice']
y=dataset['SalePrice']
```

```python
from sklearn.feature_selection import f_regression, mutual_info_regression
scores=[]
for c in list(X.columns):
    f_test, _ = f_regression(X[c].reshape(-1,1), y)
    mi = mutual_info_regression(X[c].reshape(-1,1), y)
    scores.append((c,f_test,mi))

```

```python
scores = pd.DataFrame(scores, columns=['Variable','F_score','Mutual_info'])
```

```python
f_list=scores[scores['F_score']>=201.74].Variable.tolist()
```
